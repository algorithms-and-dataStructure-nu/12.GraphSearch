```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
from typing import Hashable
```

```python
def define_directed_graph():
    G = nx.DiGraph() # 有向グラフの作成: Create a directed graph
    # 辺の追加: Add edges
    edges = [
        ('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('B', 'G'), ('C', 'D'), ('C', 'F'),
        ('D', 'E'), ('E', 'G'), ('F', 'E'), ('G', 'F')
    ]
    G.add_edges_from(edges)
    # 頂点の位置を設定: オプション
    # Set node positions (optional)
    positions = {
        'A': (0, 0), 'B': (1, 1), 'C': (1, -1),
        'D': (1, 0), 'E': (2, 0), 'F': (3, -1), 'G': (3, 1)
    }
    nx.set_node_attributes(G, positions, 'pos')
    return G
```

```python
# グラフの描画: Draw the graph
g_d = define_directed_graph()
pos = nx.get_node_attributes(g_d, 'pos')
nx.draw(g_d, pos=pos, with_labels=True, node_color='lightblue', edge_color='black');
```

## 幅優先探索: Breadth-First Search
- 開始点から近い頂点を優先的に探索するアルゴリズム
- キュー (queue) を使って探索する頂点を管理
    - 隣接頂点をキューに追加
    - キューから頂点を取り出して探索
- 一度訪れた頂点は再度訪れないようにする
----
- An algorithm that prioritizes exploring vertices close to the starting point
- Uses a queue to manage vertices to be explored
    - Add adjacent vertices to the queue
    - Remove vertices from the queue for exploration
- Ensure that once a vertex is visited, it is not visited again

```python
def bfs(graph:nx.DiGraph, start:Hashable,showQueue:bool=False)->set[tuple[Hashable,Hashable]]:
    """
    幅優先探索: Breadth-First Search (BFS) on a directed graph.

    Parameters:
    graph (nx.DiGraph): The directed graph to traverse.
    start (Hashable): The starting node for the BFS.
    """
    visited = set()
    queue = deque()
    queue.append((None,start))
    used_edges = set()    
    while queue:
        if showQueue:
            print("Queue:", [f"({prev},{node})" for prev, node in queue])
        prev_node, node = queue.popleft()
        if node not in visited:
            visited.add(node)
            if prev_node is not None:
                used_edges.add((prev_node, node))
            neighbors = list(graph.neighbors(node))
            neighbors.sort()
            for neighbor in neighbors:
                if neighbor not in visited:
                    queue.append((node, neighbor))
    return used_edges
```

```python
# 幅優先探索の実行例: Example of running BFS
# g_d = define_directed_graph()
g_d = define_directed_graph2()
bfs_edges = bfs(g_d, start='A', showQueue=True)
pos = nx.get_node_attributes(g_d, 'pos')
nx.draw(g_d, pos=pos, with_labels=True, node_color='lightblue', edge_color='black')
nx.draw_networkx_edges(g_d, pos=pos, edgelist=bfs_edges, edge_color='red', width=2);
```
