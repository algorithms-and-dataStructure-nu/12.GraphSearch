```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
def define_directed_graph():
    G = nx.DiGraph() # 有向グラフの作成: Create a directed graph
    # 辺の追加: Add edges
    edges = [
        (1, 2), (1, 3), (2, 4),(2, 5), (2, 7), (3, 4),(3, 6),
        (4, 5), (5, 7), (6, 5), (7, 6)
    ]
    G.add_edges_from(edges)
    # 頂点の位置を設定: オプション
    # Set node positions (optional)
    positions = {
        1: (0, 0), 2: (1, 1), 3: (1, -1),
        4: (1, 0), 5: (2, 0), 6: (3, -1), 7: (3, 1)
    }
    nx.set_node_attributes(G, positions, 'pos')
    return G
```

```python
g_d = define_directed_graph()
pos = nx.get_node_attributes(g_d, 'pos')
nx.draw(g_d, pos=pos, with_labels=True, node_color='lightblue', edge_color='black');
```

## 幅優先探索: Breadth-First Search
- 開始点から近い頂点を優先的に探索するアルゴリズム
- キュー (queue) を使って探索する頂点を管理
    - 隣接頂点をキューに追加
    - キューから頂点を取り出して探索
- 一度訪れた頂点は再度訪れないようにする
----
- An algorithm that prioritizes exploring vertices close to the starting point
- Uses a queue to manage vertices to be explored
    - Add adjacent vertices to the queue
    - Remove vertices from the queue for exploration
- Ensure that once a vertex is visited, it is not visited again

```python
def bfs(graph:nx.DiGraph, start:int)->set[tuple[int,int]]:
    visited = set()
    queue = deque()
    queue.append((None,start))
    used_edges = set()    
    while queue:
        prev_node, node = queue.popleft()
        if node not in visited:
            visited.add(node)
            if prev_node is not None:
                used_edges.add((prev_node, node))
            neighbors = list(graph.neighbors(node))
            neighbors.sort()
            for neighbor in neighbors:
                if neighbor not in visited:
                    queue.append((node, neighbor))
    return used_edges
```

```python
# g_d = define_directed_graph()
g_d = define_directed_graph2()
bfs_edges = bfs(g_d, start=1)
pos = nx.get_node_attributes(g_d, 'pos')
nx.draw(g_d, pos=pos, with_labels=True, node_color='lightblue', edge_color='black')
nx.draw_networkx_edges(g_d, pos=pos, edgelist=bfs_edges, edge_color='red', width=2);
```

## 深さ優先探索 (DFS): Depth-First Search
- 開始点からできるだけ深く探索するアルゴリズム
- スタック (stack) を使って探索する頂点を管理
    - 隣接頂点をスタックに追加
    - スタックから頂点を取り出して探索
- 一度訪れた頂点は再度訪れないようにする
----
- An algorithm that explores as deeply as possible from the starting point
- Uses a stack to manage vertices to be explored
    - Add adjacent vertices to the stack
    - Remove vertices from the stack for exploration
- Ensure that once a vertex is visited, it is not visited again

```python
def dfs(graph:nx.DiGraph, start:int)->set[tuple[int,int]]:
    visited = set()
    stack = []
    stack.append((None,start))
    used_edges = set()    
    while stack:
        prev_node, node = stack.pop()
        if node not in visited:
            visited.add(node)
            if prev_node is not None:
                used_edges.add((prev_node, node))
            neighbors = list(graph.neighbors(node))
            neighbors.sort(reverse=True)
            for neighbor in neighbors:
                if neighbor not in visited:
                    stack.append((node, neighbor))
    return used_edges
```

```python
# g_d = define_directed_graph()
g_d = define_directed_graph2()
dfs_edges = dfs(g_d, start=1)
pos = nx.get_node_attributes(g_d, 'pos')
nx.draw(g_d, pos=pos, with_labels=True, node_color='lightblue', edge_color='black')
nx.draw_networkx_edges(g_d, pos=pos, edgelist=dfs_edges, edge_color='red', width=2);
```

## 深さ優先探索の再帰的実装: Recursive Implementation of Depth-First Search

```python
def dfs_recursive(graph:nx.DiGraph, start:int) -> set[tuple[int,int]]:
    visited:set[int]=set()
    used_edges:set[tuple[int,int]]=set()

    def dfs_sub(node:int):
        visited.add(node)
        neighbors = list(graph.neighbors(node))
        neighbors.sort()
        for neighbor in neighbors:
            if neighbor not in visited:
                used_edges.add((node, neighbor))
                dfs_sub(neighbor)
        

    dfs_sub(start)
    return used_edges

# g_d = define_directed_graph()
g_d = define_directed_graph2()
dfs_edges = dfs_recursive(g_d, start=1)
pos = nx.get_node_attributes(g_d, 'pos')
nx.draw(g_d, pos=pos, with_labels=True, node_color='lightblue', edge_color='black')
nx.draw_networkx_edges(g_d, pos=pos, edgelist=dfs_edges, edge_color='red', width=2);
```
